```python
'IDENTITY OPERATORS'

# Identity operators

# Identity operators are used to check if two variables point to the same object in memory, they are different from equality operators which check if two variables have the same value.

a = 5 
# a = [1,2,3,4,5] 

b = 5
# b = [1,2,3,4,5]

c = a

print("a == b:", a == b)  # Equal
 
print("a is b:", a is b)  # It returns true if values are not objects and equals to same value for example for a=5 and b=5

print("a is c:", a is c)  # Identical

print("a is not b:", a is not b)  # Not identical

print("a is not c:", a is not c)  # Not identical

print("id(a) == id(b):", id(a) == id(b))  # Equal

print("id(a) == id(c):", id(a) == id(c))  # Equal

print("id(a) is not id(b):", id(a) is not id(b))  # Not identical

print("id(a) is not id(c):", id(a) is not id(c))  # Not identical


print('\nIntegers less than 256\n')

a = 1000  
b = 1000  
print("a is b: ",a is b)  # Output: False (Different memory locations)  

a = 257  
b = 257  
print("Now \na=257 \nb=257")
print("a is b :",a is b)  # Output: True (same memory locations)  
# Reason: Python caches small integers (from -5 to 256) 
print("id(a)",id(a))
print("id(b)",id(b))
print("id(a) == id(b)",id(a) == id(b))  
print("id(a) is id(b)",id(a) is id(b))  
 
c = 10  
d = 10  
print("c is d :",c is d)  # Output: True (Python caches small integers) 
print("id(c)",id(c))
print("id(d)",id(d))
print("id(c) == id(d)",id(c) == id(d))   
print("id(c) is id(d)",id(c) is id(d))   


print("\nChecking If a Variable Is None (using 'is' is preferred)\n")

x = None  
if x is None:  
    print('x is None')  # Preferred way to check for None  
```

```python
"OPERATORS PRECEDENCE"

# Operators precedence

# Parentheses: ()
# Indexing / Slicing: [], [:]
# Exponentiation: **
# Unary plus and minus: +x, -x
# Multiplication, Division, Modulus: *, /, //, %
# Addition and Subtraction: +, -
# Bitwise Shift: <<, >>
# Bitwise AND: &
# Bitwise XOR: ^
# Bitwise OR: |
# Comparisons: ==, !=, >, <, >=, <=
# Identity: is, is not
# Membership: in, not in
# Logical NOT: not
# Logical AND: and
# Logical OR: or
# Conditional Expression: if ... else ...
# Assignment and Augmented Assignment: =, +=, -=, *=, /=, //=, %=, **=, &=, |=, ^=, <<=, >>=
# Comma: ,
# Lambda: lambda

result = None
result = (3 + 4) * 2 ** 3 / (7 - 2) % 5 // 3 + 1 - 8 in [1, 2, 3] and not (result is None) or (result == 5) ^ (3 << 1) & (10 >> 1)  

print(result)


# Breakdown:  
# (3 + 4) = 7  
# (7 - 2) = 5  
# 2 ** 3 = 8  
# 7 * 8 = 56  
# 56 / 5 = 11.2  
# 11.2 % 5 = 1.2  
# 1.2 // 3 = 0.0  
# 0.0 + 1 = 1.0  
# 1.0 - 8 = -7.0  

# Check if -7.0 is in the list [1, 2, 3] -> False  
# Evaluate: False and not (result is None) -> False  
# Evaluate: (result == 5) -> False  
# Evaluate: (3 << 1) = 6 and (10 >> 1) = 5 -> 6 & 5 = 4 (binary: 110 & 101 = 100)  

# Final expression becomes:  
# False or (False ^ 4) -> True  

```
